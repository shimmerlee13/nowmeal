#### [크롤링한 이미지 객체추출]

#### 1. redpepper이미지에서 객체 끼리 조금씩 띄어있는 객체들을 저장해보자.

```python
import cv2
import numpy as np 
from matplotlib import pyplot as plt
src = cv2.imread('C:\\Users\\user\\redpepper\\redpepper24.jpg')
print(src.shape)
# 아 폴더명이나 이미지명 한글로 하면 안불러와지네 어이가 없네 ㄷㄷ

dst = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
_,th = cv2.threshold(dst, 0,255, cv2.THRESH_OTSU+cv2.THRESH_BINARY_INV)
dst = cv2.cvtColor(th, cv2.COLOR_GRAY2BGR)

_,labels, stats, centroids = cv2.connectedComponentsWithStats(th)
# print(stats)

count = 0
area_list = []
for x,y,w,h,cnt in stats: 
#     print(w*h)
#     print(cnt)
    if  (h,w) < src.shape and w<src.shape[1] and cnt>500:
        area_list.append(w*h)
        cv2.rectangle(dst,(x,y,w,h),(0,255,0),1) ## 객체인 부분만 직사각형 그림 
        count +=1
print('1차 잡음 제거후 ',count,'개 객체가 추출 되었다.')


print('===============객체 추출 중=============')
area_thres = np.min(np.array(area_list))
print('area_thres : ',area_thres)
count=0
for x,y,w,h,cnt in stats:
    if (h,w) < src.shape and (w*h)>=area_thres and w<src.shape[1]:
        # 면적이 100보다 크고 객체너비가 이미지너비보다 작을 때
        count +=1
        obj = src[y:y+h, x:x+w]  # 이미지에서 객체 추출 
        file_name = 'obj'+ str(count)+'.jpg'
        cv2.imwrite(file_name, obj) # 추출된 객체 저장 
        
print('최종',count,'개 객체가 추출 되었다.')

```

#### 3. for문으로 그냥 전체 데이터 객체추출해보자 어떻게 나오는지 확인 

```python
import cv2
import os
import numpy as np 
from matplotlib import pyplot as plt
from random import *

search = input('찾을 식재료 이미지 폴더명 ')
original_path = "C:\\Users\\user\\"+search+"\\"
obj_path = "C:\\Users\\user\\redpepper_findall_obj\\"

file_list = os.listdir(original_path)
img_list = []

for item in file_list :
    if item.find('.jpg') is not -1 :
        img_list.append(item)
total_image = len(img_list)
print('총 ',total_image,'개입니다.')

#--------------------------------------------------------------------
for img_file in img_list:
    src = cv2.imread(original_path + img_file)
    print(src.shape)
    # 아 폴더명이나 이미지명 한글로 하면 안불러와지네 어이가 없네 ㄷㄷ

    dst = cv2.cvtColor(src, cv2.COLOR_BGR2GRAY)
    _,th = cv2.threshold(dst, 0,255, cv2.THRESH_OTSU+cv2.THRESH_BINARY_INV)
    dst = cv2.cvtColor(th, cv2.COLOR_GRAY2BGR)

    _,labels, stats, centroids = cv2.connectedComponentsWithStats(th)
    # print(stats)

    count = 0
    area_list = []
    for x,y,w,h,cnt in stats: 
    #     print(w*h)
    #     print(cnt)
        if  (h,w) < src.shape and w<src.shape[1] and cnt>500:
            area_list.append(w*h)
            cv2.rectangle(dst,(x,y,w,h),(0,255,0),1) ## 객체인 부분만 직사각형 그림 
            count +=1
    print('1차 잡음 제거후 ',count,'개 객체가 추출 되었다.')


    print('===============객체 추출 중=============')
    if count !=0:
        area_thres = np.min(np.array(area_list))
        print('area_thres : ',area_thres)
        count=0
        for x,y,w,h,cnt in stats:
            if (h,w) < src.shape and (w*h)>=area_thres and w<src.shape[1]:
                 # 면적이 100보다 크고 객체너비가 이미지너비보다 작을 때
                count +=1
                obj = src[y:y+h, x:x+w]  # 이미지에서 객체 추출 
                file_name = obj_path+'obj'+ str(count)+str(randint(1, 10))+'.jpg'
                cv2.imwrite(file_name, obj) # 추출된 객체 저장 

        print('최종',count,'개 객체가 추출 되었다.')
    cv2.imwrite(obj_path+img_file, src)

print('모든 객체 추출 끝')

result_path = "C:\\Users\\user\\redpepper_findall_obj\\"

file_list = os.listdir(result_path)
img_list = []

for item in file_list :
    if item.find('.jpg') is not -1 :
        img_list.append(item)

total_image = len(img_list)
print('모든 객체 추출한 후 총 ',total_image,'개입니다.')
```
