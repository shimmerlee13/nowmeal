#### 웹페이지 가져오기

1. 우선 웹크롤링을 하기 위해서 코드를 작성할 때 도움을 받을 모듈들을 import 해야한다.

```python
from bs4 import BeautifulSoup 
from pprint import pprint
import requests
#pprint는 딕셔너리의 데이터가 긴 경우에 좀 더 보기 편하게 보여주게 도와준다.
```

2. import 가 끝났으면 requests를 이용해서 현재 '네이버 날씨' 웹페이지의 전체 소스코드를 갖고와보자.

```python
html = requests.get('https://search.naver.com/search.naver?query=서울날씨')
#웹페이지 요청을 하는 코드이다. 특정 url을 적으면 웹피이지에 대한 소스코드들을 볼 수 있다.

pprint(html.text)
#html 이라는 변수에 저장된 소스코드들 중 텍스트들을 pprint로 정렬한걸 눈으로 확인한다.
```

3. 소스코드들을 갖고왔지만 이것들은 HTML 언어로 작성된 코드들이다.
파이썬에서 이용할 수 있는 의미있는 부분들을 보려면 Parsing 작업이 필요하다.

```python
soup = BeautifulSoup(html.text, 'html.parser')
#파이썬에서 보기 좋게, 다루기 쉽게 파싱작업을 거쳐야 각 요소에 접근이 쉬워진다.
#이것을 도와주는게 beautifulsoup4 모듈이다.
```

4. 이제 파싱작업이 끝난 데이터에서 find 함수를 통해 필요한 부분들만 추출해보자.
한단계씩 대상을 좁혀간다고 생각하면 좋다.

```python
data1 = soup.find('div', {'class':'weather_box'})
# soup 모듈의 find 함수를 사용해서 data1에 값을 저장한다.
# 매개변수에는 div 태그명과 class 라는 속성의 값이 weather_box라는 녀석을 딕셔너리로 저장하는 코드이다.
# find 함수를 사용할 때 주의할 점은 같은 웹피이지 소스코드에 같은 소스가 여러가지 있으면 
# 맨 처음 탐색된것만 반환하고 나머지는 무시된다는 점이다.
```

5. 이제 속성값이 weather_box에 해당하는 코드들을 data1에 저장했으니 weather_box 내부에 있는 정보들중에서 현재 주소를 추출해보자.

(1) 현재위치

```python
address = data1.find('span', {'class':'btn_select'}).text
print('현재 위치: '+address)
#data1 변수에 저장된 정보중 span 태그명과 btn_select라는 속성값을 갖고 있는 녀석을 딕셔너리로 저장하는 코드이다.
```

(2) 현재 온도

```python
current_temp = data1.find('span',{'class': 'todaytemp'}).text
print('현재 온도: '+current_temp+'℃')
```

(3) 일 최저 온도 출력 

```python
min_temp = data1.find('span',{'class': 'min'}).text.split('˚')[0]
print('일 최저 온도 : ',min_temp+'˚')
```

(4) 일 최고 온도 출력

```python
max_temp = data1.find('span',{'class': 'max'}).text.split('˚')[0]
print('일 최고 온도 출력 : ',max_temp+'˚')
```

(5) 일 체감 온도 출력

```python
sensible_temp = data1.find('span',{'class': 'sensible'}).text.split(' ')[-1]
print('일 체감 온도 출력 :',sensible_temp)
```

(6) 일 평균기온 출력

```python
average_temp = (eval(min_temp)+eval(max_temp))/2
print('일 평균기온 출력 :', average_temp)
```

(7) 오전 강수확률

```python
morning_rainrate = data1.find('span',{'class': 'point_time morning'}).text.split(' ')[-3]
print('오전 강수확률 :', morning_rainrate)
```

(8). 오후 강수확률

```python
morning_rainrate = data1.find('span',{'class': 'point_time afternoon'}).text.split(' ')[-3]
print('오후 강수확률 :', morning_rainrate)
```
