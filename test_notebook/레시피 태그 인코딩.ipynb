```python
import pandas as pd
import numpy as np
import math
df = pd.read_csv('real_tag.csv', encoding='CP949')
df.head(10)
```

```python
print(df.shape)
```

```python
origin = pd.DataFrame(df.iloc[:,:22])
origin
```

```python
tag = pd.DataFrame(df.iloc[:,23:])
tag
```

#### [tag 1,0 데이터 만들기]

```python
tag_col = list(tag.columns)
print(len(tag_col))
for i in range(origin.shape[0]):
    row_data = origin.iloc[i,:]
    for j in range(3,22):
        for t in tag_col:
            if row_data[j]== t:
                tag.iloc[i,tag_col.index(t)]=1
print('end')
```

```python
tag.to_csv("table_tag2.csv", mode='w',encoding='CP949')
```
